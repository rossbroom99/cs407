```python
internal = v.enum_summary(df, 'actor.internal.motive')
internal = internal.drop(internal.index[0])
internal = internal.drop(10)
internal = internal[['enum', 'x']]
internal.columns = ['internal Actor Motive', 'Count']

def avg(x):
    total = internal['Count'].sum()
    return round((x / total) * 100, 1)

internal['Percentage'] = internal['Count'].apply(lambda x: avg(x))
internal = internal.sort_values(by='Percentage', ascending=False)

internal.head(20)
```

```python
df_actors_internal = v.enum_summary(df,'actor.internal.variety',by='actor')
```

```python
df_actors_external = v.enum_summary(df,'actor.external.variety',by='actor')
```

```python
df_actors_internal.drop(['n', 'freq'], axis=1, inplace=True)
df_actors_internal.columns = ['Actor Origin', 'Actor Type', 'Count']
df_External = df_actors_internal[df_actors_internal['Actor Origin'] == 'actor.External']
df_Partners = df_actors_internal[df_actors_internal['Actor Origin'] == 'actor.Partner']
df_actors_internal.drop(df_External.index, inplace= True)
df_actors_internal.drop(df_Partners.index, inplace= True)
df_Unknown = df_actors_internal[df_actors_internal['Actor Origin'] == 'actor.Unknown']
df_actors_internal.drop(df_Unknown.index, inplace=True)
df_actors_internal['Actor Origin'] = 'Internal'
df_actors_internal
```

```python
df_actors_external.drop(['n', 'freq'], axis=1, inplace=True)
df_actors_external.columns = ['Actor Origin', 'Actor Type', 'Count']
df_Internal = df_actors_external[df_actors_external['Actor Origin'] == 'actor.Internal']
df_Partners_2 = df_actors_external[df_actors_external['Actor Origin'] == 'actor.Partner']
df_actors_external.drop(df_Internal.index, inplace= True)
df_actors_external.drop(df_Partners_2.index, inplace= True)
df_Unknown_2 = df_actors_external[df_actors_external['Actor Origin'] == 'actor.Unknown']
df_actors_external.drop(df_Unknown_2.index, inplace=True)
df_actors_external['Actor Origin'] = 'External'
df_actors_external
```

```python
df_actors_combined = pd.concat([df_actors_internal, df_actors_external])
df_actors_combined.head()
```

```python
import plotly.express as px
fig_1 = px.sunburst(df_actors_combined, path=['Actor Origin', 'Actor Type'], values='Count')
fig_1.update_layout(uniformtext_minsize=12)
plot(fig_1, filename = 'Actor Origin', auto_open=True)
```
